```python
import pandas_datareader.data as web
import pandas as pd
df_kospi = web.DataReader('^KS11', 'yahoo', start='2004-01-04')
# df_kospi.describe
```

```python
peak = df_kospi['Adj Close'].rolling(252,min_periods=1).max()
drawdown = df_kospi['Adj Close']/peak - 1.0
max_dd = drawdown.rolling(252,min_periods=1).min()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9,7))
plt.subplot(211)
df_kospi['Close'].plot(label='KOSPI', title='KOSPI MDD', grid=True, legend =True)
plt.subplot(212)
drawdown.plot(c='blue', label='KOSPI DD', grid=True, legend =True)
max_dd.plot(c='red', label='KOSPI MDD', grid=True, legend =True)
plt.show()
```
